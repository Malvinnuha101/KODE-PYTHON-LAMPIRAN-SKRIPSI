```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, root_scalar
from scipy.integrate import quad
from scipy.special import iv, kv
from functools import lru_cache
from sklearn.metrics import r2_score
import scipy.stats as stats
import warnings

warnings.filterwarnings('ignore', category=RuntimeWarning)
warnings.filterwarnings('ignore', category=UserWarning)

G_ASTRO = 4.30091e-6
A0_STD_SI = 1.2e-10
A0_ARCTAN_SI = A0_STD_SI
KM2_S2_PER_KPC_TO_SI = 3.2408e-14
A0_STD_ASTRO = A0_STD_SI / KM2_S2_PER_KPC_TO_SI
A0_ARCTAN_ASTRO = A0_ARCTAN_SI / KM2_S2_PER_KPC_TO_SI
KAPPA = 7.6695

class GalaxyRotationAnalyzer:
    def __init__(self):
        self.data = self.load_observational_data()
        self.models = {
            'MOND Std': self.v_total_mond_std,
            'MOND Arctan': self.v_total_mond_arctan,
            'ΛCDM NFW': self.v_total_nfw,
            'ΛCDM Core': self.v_total_core
        }
        self.initial_params = self.set_initial_parameters()

    @staticmethod
    def load_observational_data():
        r_mw = np.concatenate([
            np.array([0.100, 0.110, 0.123, 0.133, 0.146, 0.161, 0.177, 0.195, 0.214, 0.236,
                      0.258, 0.285, 0.314, 0.345, 0.380, 0.418, 0.459, 0.505, 0.556, 0.612,
                      0.673, 0.741, 0.814, 0.895, 0.985, 1.083, 1.192, 1.311, 1.442, 1.586,
                      1.745, 1.919, 2.111, 2.323, 2.555, 2.811, 3.091]),
            np.array([3.400, 3.740, 4.114, 4.526, 4.979, 5.476, 6.024, 6.626, 7.289, 8.018,
                      8.820, 9.702, 10.672, 11.739, 12.913, 14.204, 15.625, 17.187, 18.906,
                      20.797, 22.876, 25.164, 27.680, 30.448, 33.493, 36.842, 40.527, 44.579,
                      49.037, 53.941, 59.335, 65.268, 71.795, 78.975, 86.872, 95.561])
        ])
        v_mw = np.concatenate([
            np.array([144.9, 147.4, 150.4, 153.8, 158.9, 162.4, 180.1, 196.6, 213.6, 227.8,
                      234.4, 244.1, 248.2, 250.2, 251.0, 250.7, 249.7, 248.0, 245.0, 243.2,
                      239.8, 235.3, 231.7, 227.8, 224.5, 221.7, 210.1, 216.8, 214.7, 212.7,
                      211.2, 209.5, 208.5, 208.2, 208.9, 210.4, 213.4]),
            np.array([217.2, 222.0, 226.6, 229.5, 231.6, 234.1, 237.2, 240.5, 241.1, 236.0,
                      236.7, 234.5, 234.2, 237.1, 242.8, 248.5, 240.7, 246.2, 246.5, 243.9,
                      241.6, 243.7, 237.3, 229.6, 222.5, 218.0, 207.1, 201.5, 194.2, 189.8,
                      186.2, 184.7, 183.9, 181.4, 179.5, 167.7])
        ])
        sigma_v_mw = np.concatenate([
            np.array([3.7, 4.2, 4.8, 6.1, 10.3, 16.1, 23.4, 27.1, 26.9, 22.7, 17.0,
                      11.8, 7.6, 4.7, 2.9, 2.1, 2.3, 2.9, 3.7, 4.6, 5.4, 6.7, 6.5,
                      6.0, 5.2, 4.5, 4.0, 3.7, 3.4, 3.1, 2.9, 2.3, 1.8, 1.6, 2.2, 3.6, 4.8]),
            np.array([5.9, 6.6, 5.7, 4.1, 4.3, 5.3, 5.7, 5.0, 4.1, 4.4, 5.1, 6.0,
                      7.1, 9.8, 12.4, 13.3, 14.8, 17.4, 17.3, 17.5, 15.6, 15.2, 16.1,
                      15.3, 14.1, 14.0, 13.8, 12.7, 12.9, 11.1, 10.4, 9.6, 9.3, 13.0, 14.6, 16.5])
        ])

        r_m31 = np.concatenate([
            np.array([0.100, 0.120, 0.144, 0.173, 0.207, 0.249, 0.299, 0.358, 0.430, 0.516,
                      0.619, 0.743, 0.892, 1.070, 1.284, 1.541, 1.849, 2.219, 2.662, 3.195,
                      3.834, 4.601, 5.521, 6.625, 7.950, 9.540, 11.448, 13.737]),
            np.array([16.5, 19.8, 23.7, 28.5, 34.2, 41.0, 49.2, 59.1, 70.9, 85.1,
                      102.1, 122.5, 147.0, 176.4, 211.6, 254.0, 304.8, 365.7, 438.9, 526.6,
                      632.0, 758.4, 910.0, 1092.1, 1310.5, 1572.6, 1887.1])
        ])
        v_m31 = np.concatenate([
            np.array([183.9, 190.7, 204.1, 207.6, 210.2, 213.5, 217.2, 219.7, 219.4, 216.2,
                      213.7, 219.5, 221.3, 235.3, 242.8, 248.3, 252.9, 245.0, 239.0, 236.6,
                      228.4, 222.7, 229.1, 243.2, 255.4, 261.4, 263.5, 262.8]),
            np.array([257.3, 246.1, 236.5, 232.3, 233.5, 230.4, 237.1, 249.4, 218.5, 206.6,
                      213.0, 197.4, 177.5, 165.3, 165.1, 160.8, 151.1, 122.0, 168.3, 200.5,
                      239.0, 248.1, 250.4, 241.9, 235.0, 332.7, 414.9])
        ])
        sigma_v_m31 = np.concatenate([
            np.array([0.1, 10.3, 22.6, 4.6, 3.9, 4.2, 3.8, 2.2, 2.2, 3.7, 10.9, 22.6,
                      30.7, 55.7, 57.0, 52.4, 50.6, 35.2, 16.5, 9.2, 21.1, 30.4, 29.0,
                      23.6, 18.5, 14.7, 11.4, 10.0]),
            np.array([12.2, 13.4, 11.7, 16.6, 27.4, 52.4, 109.4, 134.6, 132.4, 135.0,
                      132.5, 105.7, 81.8, 74.7, 76.9, 77.4, 74.8, 96.5, 125.7, 123.4,
                      147.8, 145.0, 141.4, 138.7, 103.5, 109.5, 69.6])
        ])

        return {
            'Bima Sakti': {'r': r_mw, 'v': v_mw, 'sigma_v': sigma_v_mw},
            'M31': {'r': r_m31, 'v': v_m31, 'sigma_v': sigma_v_m31}
        }

    @staticmethod
    def sigma_b(r, Sigma_be, a_b):
        return Sigma_be * np.exp(-KAPPA * ((r / a_b)**0.25 - 1))

    @staticmethod
    def dsigma_b_dx(x, Sigma_be, a_b):
        return GalaxyRotationAnalyzer.sigma_b(x, Sigma_be, a_b) * (-KAPPA / (4 * a_b)) * (x / a_b)**(-0.75)

    @lru_cache(maxsize=None)
    def rho_b_scalar(self, r_val, Sigma_be, a_b):
        if r_val == 0:
            integrand = lambda x: self.dsigma_b_dx(x, Sigma_be, a_b) / x
            integral, _ = quad(integrand, 1e-10, np.inf, epsabs=1e-4, epsrel=1e-4, limit=100)
        else:
            integrand = lambda x: self.dsigma_b_dx(x, Sigma_be, a_b) / np.sqrt(x**2 - r_val**2)
            integral, _ = quad(integrand, r_val, np.inf, epsabs=1e-4, epsrel=1e-4, limit=100)
        return -1 / np.pi * integral

    def rho_b(self, r, Sigma_be, a_b):
        return np.array([self.rho_b_scalar(ri, Sigma_be, a_b) for ri in np.atleast_1d(r)])

    @lru_cache(maxsize=None)
    def M_b_scalar(self, r_val, Sigma_be, a_b):
        if r_val == 0:
            return 0.0
        integrand = lambda s: 4 * np.pi * s**2 * self.rho_b_scalar(s, Sigma_be, a_b)
        mass, _ = quad(integrand, 1e-10, r_val, epsabs=1e-4, epsrel=1e-4, limit=100)
        return mass

    def M_b(self, r, Sigma_be, a_b):
        return np.array([self.M_b_scalar(ri, Sigma_be, a_b) for ri in np.atleast_1d(r)])

    def v_bulge_sq(self, r, Sigma_be, a_b):
        r_arr = np.atleast_1d(r)
        mass = self.M_b(r_arr, Sigma_be, a_b)
        v_sq = np.zeros_like(r_arr)
        mask = r_arr > 0
        v_sq[mask] = G_ASTRO * mass[mask] / r_arr[mask]
        return v_sq

    @staticmethod
    def v_disk_sq(r, Sigma_d0, a_d):
        r_arr = np.atleast_1d(r)
        v_sq = np.zeros_like(r_arr)
        mask = r_arr > 0
        r_nonzero = r_arr[mask]
        y = r_nonzero / (2 * a_d)
        bessel_term = iv(0, y) * kv(0, y) - iv(1, y) * kv(1, y)
        v_sq[mask] = np.pi * G_ASTRO * Sigma_d0 * (r_nonzero**2 / a_d) * bessel_term
        return v_sq

    @staticmethod
    def v_gas_sq(r):
        return np.zeros_like(np.atleast_1d(r))

    @staticmethod
    def v_nfw_sq(r, rho_0, h):
        r_arr = np.atleast_1d(r)
        v_sq = np.zeros_like(r_arr)
        mask = r_arr > 0
        r_nonzero = r_arr[mask]
        x = r_nonzero / h
        log_term = np.log(1 + x)
        fraction = x / (1 + x)
        v_sq[mask] = 4 * np.pi * G_ASTRO * rho_0 * h**2 * (log_term - fraction) / x
        return v_sq

    @staticmethod
    def v_core_sq(r, rho_0, h):
        r_arr = np.atleast_1d(r)
        v_sq = np.zeros_like(r_arr)
        mask = r_arr > 0
        r_nonzero = r_arr[mask]
        r3_ratio = (r_nonzero / h)**3
        log_term = np.log1p(r3_ratio)
        v_sq[mask] = (4 * np.pi * G_ASTRO / 3) * rho_0 * (h**3 / r_nonzero) * log_term
        return v_sq

    def v_baryon_sq(self, r, Sigma_be, a_b, Sigma_d0, a_d):
        return self.v_bulge_sq(r, Sigma_be, a_b) + self.v_disk_sq(r, Sigma_d0, a_d)

    def v_mond_std_sq(self, r, v_b_sq):
        r_arr = np.atleast_1d(r)
        v_b_sq_arr = np.atleast_1d(v_b_sq)
        v_m_sq = np.zeros_like(r_arr)
        mask = v_b_sq_arr > 1e-9
        v_b_sq_valid = v_b_sq_arr[mask]
        r_valid = r_arr[mask]
        term_in_sqrt = 1 + (2 * A0_STD_ASTRO * r_valid / v_b_sq_valid)**2
        v_m_sq[mask] = (v_b_sq_valid / np.sqrt(2)) * np.sqrt(1 + np.sqrt(term_in_sqrt))
        return v_m_sq

    def solve_a_m_arctan(self, a_b, tol=1e-6, verbose=False):
        def f(u):
            return u * (2 / np.pi) * np.arctan((np.pi * u) / 2) - a_b / A0_ARCTAN_ASTRO
        def f_prime(u):
            return (2 / np.pi) * np.arctan((np.pi * u) / 2) + u / (1 + ((np.pi * u) / 2)**2)
        u_guess = a_b / A0_ARCTAN_ASTRO + np.sqrt(a_b / A0_ARCTAN_ASTRO)
        u = u_guess
        if verbose:
            print("\nIterasi Newton-Raphson untuk a_M (arctan):")
            print(f"Estimasi awal u: {u:.6e}")
        converged = False
        for i in range(50):
            delta = f(u) / f_prime(u)
            if verbose:
                print(f"Iterasi {i+1}: u = {u:.6e}, delta = {delta:.6e}")
            u -= delta
            if abs(delta) < tol:
                if verbose:
                    print(f"Konvergen setelah {i+1} iterasi.")
                converged = True
                break
        if not converged:
            warnings.warn(f"\nPeringatan: Iterasi Newton-Raphson tidak konvergen setelah 50 iterasi untuk a_b = {a_b:.6e}."
                         f"Menggunakan nilai terakhir u = {u:.6e} dengan error = {abs(delta):.6e} (tol = {tol}).")
        a_m = u * A0_ARCTAN_ASTRO
        if verbose:
            print(f"a_M akhir: {a_m:.6e} (dalam satuan astro)")
        return a_m

    def v_mond_arctan_sq(self, r, v_b_sq):
        r_arr = np.atleast_1d(r)
        v_b_sq_arr = np.atleast_1d(v_b_sq)
        v_m_sq = np.zeros_like(r_arr)
        mask = v_b_sq_arr > 1e-9
        for i in np.where(mask)[0]:
            a_b = v_b_sq_arr[i] / r_arr[i]
            a_m = self.solve_a_m_arctan(a_b)
            v_m_sq[i] = a_m * r_arr[i]
        return v_m_sq

    def v_total_mond_std(self, r, Sigma_be, a_b, Sigma_d0, a_d):
        v_b_sq = self.v_baryon_sq(r, Sigma_be, a_b, Sigma_d0, a_d)
        v_total_sq = self.v_mond_std_sq(r, v_b_sq)
        return np.sqrt(np.maximum(0, v_total_sq))

    def v_total_mond_arctan(self, r, Sigma_be, a_b, Sigma_d0, a_d):
        v_b_sq = self.v_baryon_sq(r, Sigma_be, a_b, Sigma_d0, a_d)
        v_total_sq = self.v_mond_arctan_sq(r, v_b_sq)
        return np.sqrt(np.maximum(0, v_total_sq))

    def v_total_nfw(self, r, Sigma_be, a_b, Sigma_d0, a_d, rho_0, h):
        v_b2 = self.v_bulge_sq(r, Sigma_be, a_b)
        v_d2 = self.v_disk_sq(r, Sigma_d0, a_d)
        v_g2 = self.v_gas_sq(r)
        v_h2 = self.v_nfw_sq(r, rho_0, h)
        v_total_sq = v_b2 + v_d2 + v_g2 + v_h2
        return np.sqrt(np.maximum(0, v_total_sq))

    def v_total_core(self, r, Sigma_be, a_b, Sigma_d0, a_d, rho_0, h):
        v_b2 = self.v_bulge_sq(r, Sigma_be, a_b)
        v_d2 = self.v_disk_sq(r, Sigma_d0, a_d)
        v_g2 = self.v_gas_sq(r)
        v_h2 = self.v_core_sq(r, rho_0, h)
        v_total_sq = v_b2 + v_d2 + v_g2 + v_h2
        return np.sqrt(np.maximum(0, v_total_sq))

    def set_initial_parameters(self):
        return {
            'Bima Sakti': {
                'MOND Std': {'p0': [0.66e9, 1.60, 0.61e9, 8.23], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59], [0.70e9, 1.69, 0.96e9, 8.87])},
                'MOND Arctan': {'p0': [0.66e9, 1.60, 0.61e9, 8.23], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59], [0.70e9, 1.69, 0.96e9, 8.87])},
                'ΛCDM NFW': {'p0': [0.66e9, 1.60, 0.61e9, 8.23, 0.71e4, 1.07e3], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59, 0.37e4, 0.73e3], [0.70e9, 1.69, 0.96e9, 8.87, 1.05e4, 1.41e3])},
                'ΛCDM Core': {'p0': [0.66e9, 1.60, 0.61e9, 8.23, 0.71e4, 1.07e3], 'bounds': ([0.62e9, 1.51, 0.26e9, 7.59, 0.37e4, 0.73e3], [0.70e9, 1.69, 0.96e9, 8.87, 1.05e4, 1.41e3])}
            },
            'M31': {
                'MOND Std': {'p0': [0.70e9, 1.31, 0.61e9, 4.52], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63], [0.85e9, 1.62, 0.93e9, 6.41])},
                'MOND Arctan': {'p0': [0.70e9, 1.31, 0.61e9, 4.52], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63], [0.85e9, 1.62, 0.93e9, 6.41])},
                'ΛCDM NFW': {'p0': [0.70e9, 1.31, 0.61e9, 4.52, 1.70e7, 14.8], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63, 0.25e7, 7.6], [0.85e9, 1.62, 0.93e9, 6.41, 3.15e7, 22.0])},
                'ΛCDM Core': {'p0': [0.70e9, 1.31, 0.61e9, 4.52, 1.70e7, 14.8], 'bounds': ([0.55e9, 1.00, 0.29e9, 2.63, 0.25e7, 7.6], [0.85e9, 1.62, 0.93e9, 6.41, 3.15e7, 22.0])}
            }
        }

    def analyze_model(self, galaxy_name, model_name):
        data = self.data[galaxy_name]
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']
        fit_function = self.models[model_name]
        p0 = self.initial_params[galaxy_name][model_name]['p0']
        bounds = self.initial_params[galaxy_name][model_name]['bounds']

        print(f"\n{'='*80}")
        print(f"ANALISIS: Model {model_name} untuk Galaksi {galaxy_name}")
        print(f"{'='*80}")

        try:
            params, covariance = curve_fit(fit_function, r_data, v_data, p0=p0, sigma=sigma_v_data,
                                           bounds=bounds, maxfev=15000, method='trf', jac='3-point')
            errors = np.sqrt(np.diag(covariance))
            print("[SUCCESS] Fitting berhasil.")

            if 'MOND' in model_name:
                Sigma_be, a_b, Sigma_d0, a_d = params
                print("\n--- PARAMETER FITTING (MOND) ---")
                print(f"Sigma_be (M_sun/kpc^2): {Sigma_be:.3e} ± {errors[0]:.2e}")
                print(f"a_b (kpc):           {a_b:.3f} ± {errors[1]:.3f}")
                print(f"Sigma_d0 (M_sun/kpc^2): {Sigma_d0:.3e} ± {errors[2]:.2e}")
                print(f"a_d (kpc):           {a_d:.3f} ± {errors[3]:.3f}")
                if model_name == 'MOND Arctan':
                    print("\n--- NILAI a_M ARCTANGENT DARI METODE ITERASI (untuk 5 radius sampel) ---")
                    sample_indices = np.linspace(0, len(r_data)-1, 5, dtype=int)
                    v_b_sq = self.v_baryon_sq(r_data, Sigma_be, a_b, Sigma_d0, a_d)
                    for idx in sample_indices:
                        r_sample = r_data[idx]
                        a_b_sample = v_b_sq[idx] / r_sample
                        a_m_sample = self.solve_a_m_arctan(a_b_sample, verbose=True)
                        print(f"Radius {r_sample:.3f} kpc: a_b = {a_b_sample:.6e}, a_M = {a_m_sample:.6e}")
            else:
                Sigma_be, a_b, Sigma_d0, a_d, rho_0, h = params
                halo_type = 'NFW' if 'NFW' in model_name else 'Core'
                print(f"\n--- PARAMETER FITTING (ΛCDM {halo_type}) ---")
                print(f"Sigma_be (M_sun/kpc^2): {Sigma_be:.3e} ± {errors[0]:.2e}")
                print(f"a_b (kpc):           {a_b:.3f} ± {errors[1]:.3f}")
                print(f"Sigma_d0 (M_sun/kpc^2): {Sigma_d0:.3e} ± {errors[2]:.2e}")
                print(f"a_d (kpc):           {a_d:.3f} ± {errors[3]:.3f}")
                print(f"rho_0 (M_sun/kpc^3):  {rho_0:.3e} ± {errors[4]:.2e}")
                print(f"h (kpc):             {h:.3f} ± {errors[5]:.3f}")

            self.plot_fit_components(galaxy_name, model_name, params, data)
            return params, covariance

        except (RuntimeError, ValueError) as e:
            print(f"[ERROR] Fitting gagal: {e}")
            return None, None

    def calculate_transition_radius(self, params, model_name, r_max):
        Sigma_be, a_b, Sigma_d0, a_d = params
        if 'Arctan' in model_name:
            a0_val = A0_ARCTAN_ASTRO
        else:
            a0_val = A0_STD_ASTRO

        def func(r):
            if r < 1e-4: return 1e9
            v_b_sq = self.v_baryon_sq(r, Sigma_be, a_b, Sigma_d0, a_d)[0]
            g_bar = v_b_sq / r
            return g_bar - a0_val

        try:
            sol = root_scalar(func, bracket=[0.01, r_max * 1.5], method='brentq')
            return sol.root
        except ValueError:
            return None

    def plot_fit_components(self, galaxy_name, model_name, params, data):
        r_data = data['r']
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)
        fit_function = self.models[model_name]
        v_fit = fit_function(r_plot, *params)

        plt.figure(figsize=(12, 8))
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Data Observasi', capsize=3, markersize=5, zorder=5)

        if 'MOND' in model_name:
            Sigma_be, a_b, Sigma_d0, a_d = params
            v_bulge_fit = np.sqrt(self.v_bulge_sq(r_plot, Sigma_be, a_b))
            v_disk_fit = np.sqrt(self.v_disk_sq(r_plot, Sigma_d0, a_d))
            v_baryon_fit = np.sqrt(self.v_baryon_sq(r_plot, Sigma_be, a_b, Sigma_d0, a_d))
            label = 'MOND Std' if model_name == 'MOND Std' else 'MOND Arctan'
            plt.plot(r_plot, v_fit, '-', color='red', linewidth=2.5, label=f'Prediksi Total {label}')
            plt.plot(r_plot, v_baryon_fit, '--', color='blue', label='Total Baryonik')
            plt.plot(r_plot, v_bulge_fit, ':', color='orange', label='Bulge')
            plt.plot(r_plot, v_disk_fit, ':', color='green', label='Disk')

            r_trans = self.calculate_transition_radius(params, model_name, r_data.max())
            if r_trans:
                plt.axvline(x=r_trans, color='gray', linestyle='-.', alpha=0.8, linewidth=1.5)
                plt.text(r_trans * 1.05, plt.gca().get_ylim()[1]*0.25, f'  Transisi MOND\n  ($r \\approx {r_trans:.1f}$ kpc)',
                         rotation=90, verticalalignment='center', fontsize=10, color='#444444')
        else:
            Sigma_be, a_b, Sigma_d0, a_d, rho_0, h = params
            halo_type = 'NFW' if 'NFW' in model_name else 'Core'
            v_bulge_fit = np.sqrt(self.v_bulge_sq(r_plot, Sigma_be, a_b))
            v_disk_fit = np.sqrt(self.v_disk_sq(r_plot, Sigma_d0, a_d))
            v_halo_func = self.v_nfw_sq if 'NFW' in model_name else self.v_core_sq
            v_halo_fit = np.sqrt(v_halo_func(r_plot, rho_0, h))
            plt.plot(r_plot, v_fit, '-', color='red', linewidth=2.5, label=f'Prediksi Total ΛCDM {halo_type}')
            plt.plot(r_plot, v_bulge_fit, '--', color='orange', label='Bulge')
            plt.plot(r_plot, v_disk_fit, '--', color='green', label='Disk')
            plt.plot(r_plot, v_halo_fit, '--', color='blue', label=f'Halo {halo_type}')

        plt.xscale('log')
        plt.xlabel('Radius (kpc)', fontsize=14)
        plt.ylabel('Kecepatan Rotasi (km/s)', fontsize=14)
        plt.title(f'Kurva Rotasi {galaxy_name} - Model {model_name}', fontsize=16)
        plt.legend(loc='best', fontsize=12)
        plt.grid(True, which="both", ls="--", alpha=0.5)
        plt.ylim(bottom=0)
        plt.tight_layout()
        plt.show()

    @staticmethod
    def calculate_statistics(v_obs, sigma_v, v_model, k):
        residuals = v_obs - v_model
        chi_sq = np.sum((residuals / sigma_v) ** 2)
        n = len(v_obs)
        dof = n - k
        chi_sq_reduced = chi_sq / dof if dof > 0 else float('inf')
        aic = 2 * k + chi_sq
        bic = np.log(n) * k + chi_sq
        r_squared = r2_score(v_obs, v_model)
        return chi_sq, chi_sq_reduced, aic, bic, r_squared

    def compare_model_statistics(self, galaxy, results):
        print(f"\n{'='*80}")
        print(f"EVALUASI STATISTIK UNTUK {galaxy.upper()}")
        print(f"{'='*80}")
        print(f"{'Model':<20} | {'χ²':>10} | {'χ²/dof':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10}")
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}")

        data = self.data[galaxy]
        v_data, sigma_v_data = data['v'], data['sigma_v']
        model_stats = {}

        for model_name, result in results.items():
            if result['params'] is not None:
                k = len(result['params'])
                v_model = result['fit_function'](data['r'], *result['params'])
                chi2, chi2_red, aic, bic, r2 = self.calculate_statistics(v_data, sigma_v_data, v_model, k)
                model_stats[model_name] = {'aic': aic, 'bic': bic}
                print(f"{model_name:<20} | {chi2:>10.2f} | {chi2_red:>10.2f} | {aic:>10.2f} | {bic:>10.2f} | {r2:>10.4f}")

        print(f"{'-'*80}\n")

        if 'MOND Std' in model_stats and 'MOND Arctan' in model_stats:
            delta_aic = model_stats['MOND Arctan']['aic'] - model_stats['MOND Std']['aic']
            print(f"Perbandingan AIC (Arctan - Std): {delta_aic:.2f}")
            if delta_aic > 10:
                print("-> Bukti sangat kuat mendukung MOND Std.")
            elif delta_aic > 2:
                print("-> Bukti positif mendukung MOND Std.")
            elif delta_aic < -10:
                print("-> Bukti sangat kuat mendukung MOND Arctan.")
            elif delta_aic < -2:
                print("-> Bukti positif mendukung MOND Arctan.")
            else:
                print("-> Model setara.")

        if 'ΛCDM NFW' in model_stats and 'ΛCDM Core' in model_stats:
            delta_aic = model_stats['ΛCDM Core']['aic'] - model_stats['ΛCDM NFW']['aic']
            print(f"Perbandingan AIC (Core - NFW): {delta_aic:.2f}")
            if delta_aic > 10:
                print("-> Bukti sangat kuat mendukung ΛCDM NFW.")
            elif delta_aic > 2:
                print("-> Bukti positif mendukung ΛCDM NFW.")
            elif delta_aic < -10:
                print("-> Bukti sangat kuat mendukung ΛCDM Core.")
            elif delta_aic < -2:
                print("-> Bukti positif mendukung ΛCDM Core.")
            else:
                print("-> Model setara.")

    def compare_all_models_statistics(self, galaxy, results):
        print(f"\n{'='*80}")
        print(f"PERBANDINGAN KOMPREHENSIF SEMUA MODEL - {galaxy.upper()}")
        print(f"{'='*80}")

        data = self.data[galaxy]
        v_data, sigma_v_data = data['v'], data['sigma_v']

        model_stats = {}
        valid_models = {}

        for model_name, result in results.items():
            if result['params'] is not None:
                k = len(result['params'])
                v_model = result['fit_function'](data['r'], *result['params'])
                chi2, chi2_red, aic, bic, r2 = self.calculate_statistics(v_data, sigma_v_data, v_model, k)
                model_stats[model_name] = {
                    'chi2': chi2, 'chi2_red': chi2_red, 'aic': aic, 'bic': bic, 'r2': r2
                }
                valid_models[model_name] = True

        if not valid_models:
            print("Tidak ada model yang valid untuk perbandingan.")
            return

        print(f"{'Model':<20} | {'χ²':>10} | {'χ²/dof':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10} | {'Rank AIC':>8} | {'Rank BIC':>8}")
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*10}|{'-'*10}")

        sorted_aic = sorted(model_stats.items(), key=lambda x: x[1]['aic'])
        sorted_bic = sorted(model_stats.items(), key=lambda x: x[1]['bic'])

        aic_ranks = {model: i+1 for i, (model, _) in enumerate(sorted_aic)}
        bic_ranks = {model: i+1 for i, (model, _) in enumerate(sorted_bic)}

        for model_name, stats in model_stats.items():
            print(f"{model_name:<20} | {stats['chi2']:>10.2f} | {stats['chi2_red']:>10.2f} | "
                  f"{stats['aic']:>10.2f} | {stats['bic']:>10.2f} | {stats['r2']:>10.4f} | "
                  f"{aic_ranks[model_name]:>8} | {bic_ranks[model_name]:>8}")

        best_aic_model, best_aic_stats = sorted_aic[0]
        best_bic_model, best_bic_stats = sorted_bic[0]

        print(f"\n--- HASIL PERBANDINGAN MODEL TERBAIK ---")
        print(f"Berdasarkan AIC: {best_aic_model}")
        print(f"  - AIC: {best_aic_stats['aic']:.2f}")
        print(f"  - ΔAIC terkecil: Model ini memiliki AIC terendah")

        print(f"Berdasarkan BIC: {best_bic_model}")
        print(f"  - BIC: {best_bic_stats['bic']:.2f}")
        print(f"  - ΔBIC terkecil: Model ini memiliki BIC terendah")

        print(f"\n--- PERBANDINGAN RELATIF TERHADAP MODEL TERBAIK ---")
        print(f"{'Model':<20} | {'ΔAIC':>10} | {'ΔBIC':>10} | {'Kekuatan Bukti':>15}")
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*17}")

        for model_name, stats in model_stats.items():
            delta_aic = stats['aic'] - best_aic_stats['aic']
            delta_bic = stats['bic'] - best_bic_stats['bic']

            if delta_aic == 0:
                evidence = "TERBAIK"
            elif delta_aic <= 2:
                evidence = "Setara"
            elif delta_aic <= 6:
                evidence = "Sedang"
            elif delta_aic <= 10:
                evidence = "Lemah"
            else:
                evidence = "Tidak mendukung"

            print(f"{model_name:<20} | {delta_aic:>10.2f} | {delta_bic:>10.2f} | {evidence:>15}")

        print(f"\n--- ANALISIS CHI-SQUARE REDUCED ---")
        chi2_red_sorted = sorted(model_stats.items(), key=lambda x: abs(x[1]['chi2_red'] - 1))
        best_chi2_red_model, best_chi2_red_stats = chi2_red_sorted[0]

        print(f"Model dengan χ²/dof terdekat ke 1: {best_chi2_red_model}")
        print(f"  - χ²/dof: {best_chi2_red_stats['chi2_red']:.2f}")

        for model_name, stats in model_stats.items():
            deviation = abs(stats['chi2_red'] - 1)
            quality = "Sangat baik" if deviation <= 0.5 else "Baik" if deviation <= 1.0 else "Cukup" if deviation <= 2.0 else "Kurang"
            print(f"  {model_name:<18}: {stats['chi2_red']:.2f} ({quality})")

        print(f"\n--- REKOMENDASI MODEL ---")

        if best_aic_model == best_bic_model:
            print(f"✓ MODEL TERBAIK: {best_aic_model}")
            print(f"  - Konsisten berdasarkan AIC dan BIC")
            print(f"  - AIC: {best_aic_stats['aic']:.2f}, BIC: {best_aic_stats['bic']:.2f}")
            print(f"  - χ²/dof: {best_aic_stats['chi2_red']:.2f}, R²: {best_aic_stats['r2']:.4f}")
        else:
            print(f"✓ REKOMENDASI BERDASARKAN AIC: {best_aic_model}")
            print(f"  - AIC: {best_aic_stats['aic']:.2f}, ΔAIC: +0.00")
            print(f"  - χ²/dof: {best_aic_stats['chi2_red']:.2f}, R²: {best_aic_stats['r2']:.4f}")

            print(f"✓ REKOMENDASI BERDASARKAN BIC: {best_bic_model}")
            print(f"  - BIC: {best_bic_stats['bic']:.2f}, ΔBIC: +0.00")
            print(f"  - χ²/dof: {best_bic_stats['chi2_red']:.2f}, R²: {best_bic_stats['r2']:.4f}")

            aic_params = len(results[best_aic_model]['params'])
            bic_params = len(results[best_bic_model]['params'])

            if bic_params < aic_params:
                print(f"✓ SARAN: {best_bic_model} lebih parsimonious (lebih sedikit parameter)")
            else:
                print(f"✓ SARAN: {best_aic_model} memberikan fit lebih baik dengan parameter tambahan yang justified")

        print(f"{'='*80}\n")

    def plot_model_comparison(self, galaxy, results):
        data = self.data[galaxy]
        r_data = data['r']
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)

        plt.figure(figsize=(12, 8))
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Observasi', capsize=3, markersize=5)

        if 'ΛCDM NFW' in results and results['ΛCDM NFW']['params'] is not None:
            v_nfw = self.v_total_nfw(r_plot, *results['ΛCDM NFW']['params'])
            plt.plot(r_plot, v_nfw, 'b--', label='ΛCDM NFW')

        if 'ΛCDM Core' in results and results['ΛCDM Core']['params'] is not None:
            v_core = self.v_total_core(r_plot, *results['ΛCDM Core']['params'])
            plt.plot(r_plot, v_core, 'g:', label='ΛCDM Core')

        plt.xscale('log')
        plt.xlabel('Radius (kpc)')
        plt.ylabel('Kecepatan Rotasi (km/s)')
        plt.title(f'Perbandingan ΛCDM NFW vs Core untuk {galaxy}')
        plt.legend()
        plt.grid(True, which='both')
        plt.ylim(bottom=0)
        plt.show()

        plt.figure(figsize=(12, 8))
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Observasi', capsize=3, markersize=5)

        if 'MOND Std' in results and results['MOND Std']['params'] is not None:
            params = results['MOND Std']['params']
            v_std = self.v_total_mond_std(r_plot, *params)
            plt.plot(r_plot, v_std, 'r-', label='MOND Std')
            r_trans = self.calculate_transition_radius(params, 'MOND Std', r_data.max())
            if r_trans:
                plt.axvline(x=r_trans, color='red', linestyle=':', alpha=0.5)
                plt.text(r_trans * 1.05, 40, 'Transisi (Std)', color='red', rotation=90, fontsize=8)

        if 'MOND Arctan' in results and results['MOND Arctan']['params'] is not None:
            params = results['MOND Arctan']['params']
            v_arctan = self.v_total_mond_arctan(r_plot, *params)
            plt.plot(r_plot, v_arctan, 'm-.', label='MOND Arctan')
            r_trans = self.calculate_transition_radius(params, 'MOND Arctan', r_data.max())
            if r_trans:
                plt.axvline(x=r_trans, color='magenta', linestyle=':', alpha=0.5)
                plt.text(r_trans * 1.05, 80, 'Transisi (Arctan)', color='magenta', rotation=90, fontsize=8)

        plt.xscale('log')
        plt.xlabel('Radius (kpc)')
        plt.ylabel('Kecepatan Rotasi (km/s)')
        plt.title(f'Perbandingan MOND Std vs Arctan untuk {galaxy}')
        plt.legend()
        plt.grid(True, which='both')
        plt.ylim(bottom=0)
        plt.show()

    def plot_all_models_comparison(self, galaxy, results):
        data = self.data[galaxy]
        r_data = data['r']
        r_plot = np.logspace(np.log10(max(1e-2, r_data.min())), np.log10(r_data.max()), 200)

        plt.figure(figsize=(12, 8))
        plt.errorbar(r_data, data['v'], yerr=data['sigma_v'], fmt='ko', label='Data Observasi', capsize=3, markersize=5)

        styles = {
            'MOND Std': {'color': 'red', 'linestyle': '-'},
            'MOND Arctan': {'color': 'magenta', 'linestyle': '-.'},
            'ΛCDM NFW': {'color': 'blue', 'linestyle': '--'},
            'ΛCDM Core': {'color': 'green', 'linestyle': ':'}
        }

        for model_name, result in results.items():
            if result['params'] is not None:
                v_fit = result['fit_function'](r_plot, *result['params'])
                plt.plot(r_plot, v_fit, label=model_name, **styles[model_name])
                if 'MOND' in model_name:
                    r_trans = self.calculate_transition_radius(result['params'], model_name, r_data.max())
                    if r_trans:
                        color = styles[model_name]['color']
                        plt.axvline(x=r_trans, color=color, linestyle=':', alpha=0.4, linewidth=1)
                        y_pos = 20 if 'Std' in model_name else 50
                        trans_label = 'Transisi (Std)' if 'Std' in model_name else 'Transisi (Arctan)'
                        plt.text(r_trans * 1.05, y_pos, trans_label, color=color, rotation=90, fontsize=8)

        plt.xscale('log')
        plt.xlabel('Radius (kpc)', fontsize=14)
        plt.ylabel('Kecepatan Rotasi (km/s)', fontsize=14)
        plt.title(f'Perbandingan Semua Model untuk {galaxy}', fontsize=16)
        plt.legend(loc='best', fontsize=12)
        plt.grid(True, which="both", ls="--", alpha=0.5)
        plt.ylim(bottom=0)
        plt.tight_layout()
        plt.show()

    def plot_residuals(self, galaxy, results):
        data = self.data[galaxy]
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']

        fig, axes = plt.subplots(len(results), 1, figsize=(12, 10), sharex=True)
        if len(results) == 1:
            axes = [axes]
        fig.suptitle(f'Residual Analisis untuk {galaxy}', fontsize=16)

        colors = {'MOND Std': 'red', 'MOND Arctan': 'magenta', 'ΛCDM NFW': 'blue', 'ΛCDM Core': 'green'}

        for i, (model_name, result) in enumerate(results.items()):
            if result['params'] is not None:
                v_model = result['fit_function'](r_data, *result['params'])
                residuals = v_data - v_model
                axes[i].errorbar(r_data, residuals, yerr=sigma_v_data, fmt='o',
                                color=colors[model_name], ecolor='lightgray', capsize=3)
                axes[i].axhline(0, color='black', linestyle='--')
                axes[i].set_ylabel('Residual (km/s)')
                axes[i].set_title(model_name)
                axes[i].grid(True)

        plt.xscale('log')
        plt.xlabel('Radius (kpc)')
        plt.tight_layout()
        plt.show()

    def compare_models(self, galaxy, model1, model2, results):
        data = self.data[galaxy]
        r_data, v_data, sigma_v_data = data['r'], data['v'], data['sigma_v']

        if model1 not in results or model2 not in results:
            print(f"Data tidak tersedia untuk {model1} vs {model2}")
            return

        params1 = results[model1]['params']
        params2 = results[model2]['params']
        if params1 is None or params2 is None:
            print(f"Fitting gagal untuk {model1} atau {model2}. Perbandingan dilewati.")
            return

        func1 = results[model1]['fit_function']
        func2 = results[model2]['fit_function']

        v_pred1 = func1(r_data, *params1)
        v_pred2 = func2(r_data, *params2)

        k1 = len(params1)
        k2 = len(params2)

        chi_sq1, _, aic1, bic1, r2_1 = self.calculate_statistics(v_data, sigma_v_data, v_pred1, k1)
        chi_sq2, _, aic2, bic2, r2_2 = self.calculate_statistics(v_data, sigma_v_data, v_pred2, k2)

        delta_chi_sq = chi_sq1 - chi_sq2
        delta_aic = aic1 - aic2
        delta_bic = bic1 - bic2

        print(f"\n{'='*80}")
        print(f"PERBANDINGAN {galaxy.upper()}: {model1} vs {model2}")
        print(f"{'='*80}")
        print(f"{'Model':<20} | {'χ²':>10} | {'AIC':>10} | {'BIC':>10} | {'R²':>10}")
        print(f"{'-'*20}-|{'-'*12}|{'-'*12}|{'-'*12}|{'-'*12}")

        print(f"{model1:<20} | {chi_sq1:>10.2f} | {aic1:>10.2f} | {bic1:>10.2f} | {r2_1:>10.4f}")
        print(f"{model2:<20} | {chi_sq2:>10.2f} | {aic2:>10.2f} | {bic2:>10.2f} | {r2_2:>10.4f}")

        print("\nPerbedaan:")
        print(f"Δχ² ({model1} - {model2}): {delta_chi_sq:.2f}")
        print(f"ΔAIC ({model1} - {model2}): {delta_aic:.2f}")
        print(f"ΔBIC ({model1} - {model2}): {delta_bic:.2f}")

        print("\nInterpretasi Berdasarkan AIC:")
        if delta_aic > 10:
            print(f"  Bukti sangat kuat {model2} lebih baik.")
        elif delta_aic > 6:
            print(f"  Bukti kuat {model2} lebih baik.")
        elif delta_aic > 2:
            print(f"  Bukti sedang {model2} lebih baik.")
        elif delta_aic > 0:
            print(f"  Bukti lemah {model2} lebih baik.")
        elif delta_aic < -10:
            print(f"  Bukti sangat kuat {model1} lebih baik.")
        elif delta_aic < -6:
            print(f"  Bukti kuat {model1} lebih baik.")
        elif delta_aic < -2:
            print(f"  Bukti sedang {model1} lebih baik.")
        elif delta_aic < 0:
            print(f"  Bukti lemah {model1} lebih baik.")
        else:
            print("  Kedua model setara berdasarkan AIC.")

        df = abs(k1 - k2)
        p_value = stats.chi2.sf(abs(delta_chi_sq), df) if df > 0 else 1.0

        print(f"\nUji Signifikansi Chi-kuadrat:")
        print(f"P-value: {p_value:.4f}")

        if p_value < 0.05:
            better_model = model1 if delta_chi_sq < 0 else model2
            print(f"  Perbedaan signifikan (p < 0.05). Model terbaik: {better_model}")
        else:
            print("  Tidak ada perbedaan signifikan (p ≥ 0.05)")

        print(f"{'='*80}\n")

    def run_analysis(self):
        print("Memulai analisis kurva rotasi galaksi...")
        all_results = {}

        for galaxy in ['Bima Sakti', 'M31']:
            print(f"\n{'#' * 50}")
            print(f" MEMPROSES GALAKSI: {galaxy.upper()} ")
            print(f"{'#' * 50}")

            galaxy_results = {}
            for model_name in self.models.keys():
                params, cov = self.analyze_model(galaxy, model_name)
                galaxy_results[model_name] = {
                    'params': params,
                    'covariance': cov,
                    'fit_function': self.models[model_name]
                }

            all_results[galaxy] = galaxy_results

            self.compare_models(galaxy, 'ΛCDM NFW', 'MOND Std', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM Core', 'MOND Std', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM NFW', 'ΛCDM Core', galaxy_results)
            self.compare_models(galaxy, 'MOND Std', 'MOND Arctan', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM NFW', 'MOND Arctan', galaxy_results)
            self.compare_models(galaxy, 'ΛCDM Core', 'MOND Arctan', galaxy_results)

            if any(res['params'] is not None for res in galaxy_results.values()):
                print(f"\n--- VISUALISASI DAN STATISTIK UNTUK {galaxy.upper()} ---")
                self.compare_model_statistics(galaxy, galaxy_results)

                self.compare_all_models_statistics(galaxy, galaxy_results)

                self.plot_model_comparison(galaxy, galaxy_results)

                self.plot_all_models_comparison(galaxy, galaxy_results)

                self.plot_residuals(galaxy, galaxy_results)

            else:
                print(f"\n[INFO] Tidak ada model yang berhasil difitting untuk {galaxy}. Visualisasi dilewati.")

        print("\nAnalisis selesai!")

if __name__ == "__main__":
    analyzer = GalaxyRotationAnalyzer()
    analyzer.run_analysis()
```
